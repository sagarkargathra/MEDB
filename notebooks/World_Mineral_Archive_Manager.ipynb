```python
def get_archive_status():
    conn = sqlite3.connect(DB_PATH)
    
    # 1. Get Table Names
    tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
    
    print("ðŸ“‚ --- WORLD MINERAL ARCHIVE DASHBOARD --- ðŸ“‚")
    for table in tables['name']:
        if table == 'sqlite_sequence': continue
        
        # 2. Get row counts for each table
        count = conn.execute(f"SELECT COUNT(*) FROM {table}").fetchone()[0]
        print(f"ðŸ”¹ Table: {table:<15} | Rows: {count:,}")
    
    conn.close()

get_archive_status()
```

```python
import sqlite3
import pandas as pd

DB_PATH = "World_Mineral_Archive.db"

def inspect_lithium_raw():
    conn = sqlite3.connect(DB_PATH)
    
    # Selecting ALL columns for Lithium to see what's hidden in the sub-commodities
    query = """
    SELECT * FROM BGS_Global 
    WHERE bgs_commodity_trans LIKE '%Lithium%' 
    AND year_clean = '1983'
    ORDER BY quantity DESC
    """
    
    df = pd.read_sql(query, conn)
    conn.close()
    return df

# This will show you the wide view with all the technical columns
df_lithium = inspect_lithium_raw()
df_lithium
```

```python
import sqlite3
import pandas as pd

DB_PATH = "World_Mineral_Archive.db"

def get_clean_lithium_report():
    conn = sqlite3.connect(DB_PATH)
    
    # Applying your specific renaming structure
    query = """
    SELECT 
        year_clean AS year,
        country_trans AS country,
        country_iso2_code AS iso_code,
        bgs_statistic_type_trans AS statistic_type,
        bgs_commodity_trans AS commodity,
        bgs_sub_commodity_trans AS sub_commodity,
        quantity,
        units,
        concat_figure_notes_text AS foot_notes
    FROM BGS_Global 
    WHERE LOWER(bgs_commodity_trans) LIKE '%lithium%'
    AND LOWER(bgs_statistic_type_trans) = 'production'
    ORDER BY year ASC, country ASC
    """
    
    df = pd.read_sql(query, conn)
    conn.close()
    return df

# To view the renamed data:
df_lithium_final = get_clean_lithium_report()
display(df_lithium_final)
```
